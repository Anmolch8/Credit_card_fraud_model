```python
import pandas as pd
```

### droping columns


```python
data = pd.read_csv('data/card_transdata.csv')
data.head()
```

```python
data.drop(columns=["ratio_to_median_purchase_price"],inplace=True)
```

```python
data.head()
```

```python
num_features=[feature for feature in data.columns if data[feature].dtype !='o']
cat_features=[feature for feature in data.columns if feature not in num_features]
```

### handling null values


```python
for feature in num_features:
    if data[feature].isna().sum() != 0:
        data[feature].fillna(data[feature].mean(),inplace=True)

data[num_features].isna().sum()
```

```python
for feature in data.iloc[:,2:].columns:
    data[feature]=data[feature].astype(int)
 
data
```

### storing new data in csv

```python
data.to_csv("data/cleansed_data.csv",index=False)
```
